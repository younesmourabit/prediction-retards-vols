```python
import pandas as pd
import numpy as np

# Fonction de conversion sécurisée
def time_to_minutes(time_str):
    """
    Convertit un string 'HH:MM' en minutes depuis minuit.
    Gère les valeurs manquantes et les formats invalides.
    """
    if pd.isna(time_str):
        return np.nan  # Conserve les NaN
    try:
        hours, minutes = map(int, str(time_str).split(':'))
        return hours * 60 + minutes
    except (ValueError, AttributeError):
        return np.nan  # En cas de format invalide (ex: "12.30" au lieu de "12:30")

# Liste des colonnes à convertir (à adapter selon votre dataset)
time_columns = [
    'CRS_DEP_TIME', 
    'DEP_TIME', 
    'WHEELS_OFF', 
    'WHEELS_ON', 
    'CRS_ARR_TIME', 
    'ARR_TIME'
]

# Application à toutes les colonnes
for col in time_columns:
    new_col_name = f"{col}_MIN"  # Nouveau nom de colonne
    df[new_col_name] = df[col].apply(time_to_minutes)

# Optionnel : Supprimer les colonnes originales
# df = df.drop(columns=time_columns)

# Aperçu des résultats
print(df[['CRS_DEP_TIME', 'CRS_DEP_TIME_MIN']].head())
```

```python
# --------------------------------------------
# 1. Importation des bibliothèques
# --------------------------------------------
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

# --------------------------------------------
# 2. Chargement des données
# --------------------------------------------
# Afficher les premières lignes pour vérification
print("Données brutes :")
print(df.head())

# --------------------------------------------
# 3. Prétraitement des données
# --------------------------------------------
# Clipper les retards négatifs à 0
df["ARR_DELAY"] = df["ARR_DELAY"].clip(lower=0)

# Encodage manuel des variables manquantes
def custom_target_encoder(df, col, target):
    if col in df.columns:
        mean = df.groupby(col)[target].mean().to_dict()
        df[f"{col}_ENCODED"] = df[col].map(mean)
        df = df.drop(columns=[col])
    return df

df = custom_target_encoder(df, "AIRLINE", "ARR_DELAY")
df = custom_target_encoder(df, "ORIGIN", "ARR_DELAY")
df = custom_target_encoder(df, "DEST", "ARR_DELAY")

# Suppression des colonnes problématiques
colonnes_a_supprimer = [
    "DEP_DELAY", "TAXI_IN", "WHEELS_OFF_MIN",
    "WHEELS_ON_MIN", "ARR_TIME_MIN", "DIVERTED"
]
df = df.drop(columns=[col for col in colonnes_a_supprimer if col in df.columns])

# Gestion des valeurs manquantes
df.fillna({
    'AIRLINE_ENCODED': df['AIRLINE_ENCODED'].mean(),
    'ORIGIN_ENCODED': df['ORIGIN_ENCODED'].mean(),
    'DEST_ENCODED': df['DEST_ENCODED'].mean()
}, inplace=True)

# --------------------------------------------
# 4. Séparation des features/cible
# ----------------------------

```

```python
from sklearn.cluster import KMeans

# Features pour le clustering (ex: durée du retard, saison, compagnie)
cluster_features = df[['ARR_DELAY', 'MONTH', 'AIRLINE_ENCODED', 'ORIGIN_ENCODED']]

kmeans = KMeans(n_clusters=5)  # 5 causes hypothétiques
df['INFERRED_CAUSE_CLUSTER'] = kmeans.fit_predict(cluster_features)

# Étiqueter manuellement les clusters basés sur les centroïdes
cluster_labels = {
    0: "Problème compagnie",
    1: "Météo",
    2: "Trafic aérien",
    3: "Retour au hangar",
    4: "Autre"
}
df['INFERRED_CAUSE'] = df['INFERRED_CAUSE_CLUSTER'].map(cluster_labels)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.multioutput import MultiOutputClassifier, MultiOutputRegressor
from xgboost import XGBClassifier, XGBRegressor
```

```python
df['RETARD_BINAIRE'] = (df['ARR_DELAY'] > 0).astype(int)  # Classification binaire
df['DUREE_RETARD'] = df['ARR_DELAY']  # Régression
```

```python
features = ['TAXI_OUT', 'AIR_TIME', 'DISTANCE', 'DAY', 'MONTH', 'YEAR',
            'CRS_DEP_TIME_MIN', 'DEP_TIME_MIN', 'CRS_ARR_TIME_MIN',
            'ORIGIN_ENCODED', 'AIRLINE_ENCODED', 'DEST_ENCODED']
```

```python
numeric_features = ['TAXI_OUT', 'AIR_TIME', 'DISTANCE', 'DAY', 'MONTH', 'YEAR',
                   'CRS_DEP_TIME_MIN', 'DEP_TIME_MIN', 'CRS_ARR_TIME_MIN']
encoded_features = ['ORIGIN_ENCODED', 'AIRLINE_ENCODED', 'DEST_ENCODED']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('enc', 'passthrough', encoded_features)
    ])
```

```python
pipeline_clf = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', XGBClassifier())
])
```

```python
pipeline_reg = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', XGBRegressor())
])
```

```python
y_class = df['RETARD_BINAIRE']
y_reg = df['DUREE_RETARD']
```

```python
# 2. Pipeline de prétraitement
# --------------------------------------------
numeric_features = ['TAXI_OUT', 'AIR_TIME', 'DISTANCE', 'DAY', 'MONTH', 'YEAR',
                   'CRS_DEP_TIME_MIN', 'DEP_TIME_MIN', 'CRS_ARR_TIME_MIN']
encoded_features = ['ORIGIN_ENCODED', 'AIRLINE_ENCODED', 'DEST_ENCODED']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('enc', 'passthrough', encoded_features)
    ])

# --------------------------------------------
# 3. Modèle multi-tâches
# --------------------------------------------
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', MultiOutputRegressor(XGBRegressor())),
    ('classifier', MultiOutputClassifier(XGBClassifier()))
])

# Définition des cibles
X = df[features]
y = df[['RETARD_BINAIRE', 'DUREE_RETARD', 'INFERRED_CAUSE_CLUSTER']]
```

```python
X_train, X_test, y_class_train, y_class_test = train_test_split(X, y_class, test_size=0.2, random_state=42)
_, _, y_reg_train, y_reg_test = train_test_split(X, y_reg, test_size=0.2, random_state=42)
```

```python
pipeline_clf.fit(X_train, y_class_train)
pipeline_reg.fit(X_train, y_reg_train)
```

```python
y_pred_class = pipeline_clf.predict(X_test)
y_pred_reg = pipeline_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, mean_squared_error

acc = accuracy_score(y_class_test, y_pred_class)
mse = mean_squared_error(y_reg_test, y_pred_reg)

print(f"Accuracy (RETARD_BINAIRE) : {acc:.2f}")
print(f"MSE (DUREE_RETARD) : {mse:.2f}")
from math import sqrt
rmse = sqrt(198.02)  # RMSE : Root Mean Squared Error
print(f"RMSE : {rmse:.2f} minutes")  # ~14.07 minutes

```

```python
from sklearn.preprocessing import LabelEncoder

# Encoder la cause du retard
cause_encoder = LabelEncoder()
df['CAUSE_ENCODED'] = cause_encoder.fit_transform(df['INFERRED_CAUSE'])
```

```python
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier

X_cause = df[df['RETARD_BINAIRE'] == 1][features]
y_cause = df[df['RETARD_BINAIRE'] == 1]['CAUSE_ENCODED']

pipeline_cause = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', XGBClassifier())
])

pipeline_cause.fit(X_cause, y_cause)
```

```python
def predire_retard(info_vol):
    import pandas as pd

    input_df = pd.DataFrame([info_vol])

    # Prédiction du retard (0 ou 1)
    retard_binaire = pipeline_clf.predict(input_df)[0]

    if retard_binaire == 0:
        print("✅ Ce vol ne devrait pas avoir de retard.")
        return {
            'retard': False,
            'duree_estimee': 0,
            'cause_probable': None
        }

    # Prédiction de la durée du retard
    duree_retard = pipeline_reg.predict(input_df)[0]

    # Prédiction de la cause du retard
    cause_pred_encoded = pipeline_cause.predict(input_df)[0]
    cause_pred = cause_encoder.inverse_transform([cause_pred_encoded])[0]

    print("⚠️ Ce vol risque d'avoir un retard.")
    print(f"⏱️ Durée estimée : {duree_retard:.1f} minutes")
    print(f"❓ Cause probable : {cause_pred}")

    return {
        'retard': True,
        'duree_estimee': duree_retard,
        'cause_probable': cause_pred
    }
```

Taxi-out : Temps que l'avion met pour rouler au sol avant le décollage.

Air time : Temps que l'avion passe en vol, entre le décollage et l'atterrissage.

Distance : Longueur du trajet entre l'aéroport de départ et celui d'arrivée.

CRS DEP TIME : Heure prévue de départ selon l'horaire du vol.

DEP TIME : Heure réelle à laquelle l'avion quitte l'aéroport.

CRS ARR TIME : Heure prévue d'arrivée à destination.

ARR TIME : Heure réelle d'atterrissage à destination.

Origin Encoded : Code numérique représentant l'aéroport de départ.

Airline Encoded : Code numérique représentant la compagnie aérienne.

Dest Encoded : Code numérique représentant l'aéroport de destination.
